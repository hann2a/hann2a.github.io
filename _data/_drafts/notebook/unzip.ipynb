```python
import sys
from io import StringIO
input_data ="""\
10
3
A 10
B 7
C 5
4
B 20
C 19
E 18
R 17
5
Y 10
J 3
G 5
Z 7
A 11
6
Q 16
P 5
O 1
I 2
B 3
V 4
7
C 6
Z 20
M 10
J 10
G 10
H 8
T 2
8
R 1
Y 3
U 9
L 15
Z 14
N 11
X 8
W 7
10
S 4
D 12
H 3
A 16
O 9
X 5
V 1
C 5
H 10
W 14
1
U 14
2
I 6
K 2
10
S 4
D 12
H 13
A 16
O 19
X 15
V 11
C 15
H 11
W 13"""

sys.stdin = StringIO(input_data)

T = int(sys.stdin.readline())

for test_case in range(1, T+1):
    case = int(sys.stdin.readline())
    alphabet, count = sys.stdin.readline().split()
    count = int(count)
    content = [ alphabet * count for _ in range(case)]

    count_alpha = 0 
    result = list()

    for alpha in content:
        result.append(alpha)
        count_alpha += 1 
        if count_alpha == 10:
            result.append("\n")

    result_str = ''.join(result)
    print(f"{T} {result_str}")
```

```python
T = int(input())

for test_case in range(1, T+1):
    case = int(input()) # 알파벳의 개수 

    collection = []

    for _ in range(case):
        alphabet, count = input().split()  # A 10 
        pair = (alphabet, count)
        collection.append(pair)
        
    # 전부 다 붙어 있는 문자열을 만들기 
    content = ''
    for alphabet, count in collection:
            content += alphabet * int(count) # AAAAAAAAAA
    
    # 10개씩 끊어서 출력할 형태 만들기 
    count_alpha = 0 # 10개씩 셀 count 
    result_str = '' #\n을 포함할 문자열 
    for str in content: # 문자열의 하나씩 출력하기 
        result_str += str
        count_alpha += 1 
        if count_alpha % 10 == 0:
            result_str += '\n'

    print(f"#{test_case}\n{result_str}")
```
