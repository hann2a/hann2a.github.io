```python
numbers = [64, 13, 9, 62, 3]
def bubble_sort(arr, N): #배열과 원소의 개수 
    for i in range(N-1, 0, -1): # 배열의 끝에서부터 1씩 줄어듦(가장 마지막 배열부터 정렬)
        for j in range(i): # 비교할 원소의 인덱스 
            if arr[j] > arr[j+1]: # 만약 왼쪽이 오른쪽보다 크다면 
                arr[j], arr[j+1] = arr[j+1], arr[j] # 서로 바꿈 
    return arr
    
sorted_numbers = bubble_sort(numbers, len(numbers))
print("정렬 후:", sorted_numbers)
```
