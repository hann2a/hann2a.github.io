```python
"""
- 안전하게 구간 간격 검사 `possible` 과 그리디 이동 `while now_position + K < N` 을 나누어서 구현
- 충전소 존재 여부 배열 `stop_arr`이 0과 1로 이루어져 있어서 `True`와 `False`로 사용
- 충전소가 있으면 해당 인덱스를 현재 위치로 전환
"""
"""
충전으로 조금 밖에 못감 (k만큼)
0에서 출발 N까지 이동  
정류장 M개(번호 줌)
최소한 몇 번의 충전을 해야 종점에 갈 수 있는가? 
충전기 설치 잘못돼서 도착 못하면 0 출력 
출발지에는 항상 충전기가 있고 충전횟수 미포함 

K: 충전하고 갈 수 있는 정류장의 수 
N: 종점 
M: 충전기 수 
"""
T = int(input())
for test_case in range(1, 1 + T):
    K, N, M = map(int, input().split())
    bus_stop = list(map(int, input().split()))

    # 구간 간격 검사 하기 
    # 충전소와 충전소, 충전소와 종점까지의 거리 > 가능한 거리? 
    # 시작점 
    # prev: 왼쪽에 있는 점 
    # possible: 
    prev = 0 
    possible = True
    for stop in bus_stop + [N]:          # 종점 + 버스 정류장 위치 인덱스
        if stop - prev > K:
            possible = False
            break
        prev = stop
	if possible is not True:
    # if not possible:
        print(f"#{test_case} 0")
        continue                         # 여기서 continue는 이 밖의 for문인 
                                         # 다음 test_case로 넘어가게 함 
    # ──────────────────────────────────────────────

    # 충전소 존재 여부 배열
    # 시작점도 세므로 + 1

    # 버스정류장 유무를 0과 1로 표현 
    stop_arr = [0] * (N + 1)
    for stop in bus_stop:
        stop_arr[stop] = 1

    # 실제 이동 & 충전 횟수 카운트
    # now_position: 이동하면서 저장할 현재 위치 
    # charge_count: 충전 몇 번 했는지 세기 

    now_position = 0
    charge_count = 0
    # else문: 범위 내에 없었으면 charge_count
    while now_position + K < N:              # 종점에 도달할 때까지
        # K 이내에서 가장 멀리 있는 충전소 탐색
        # range(지금위치+ 갈 수 있는 만큼, 지금 위치, -1)
        # 따라서 제일 좋은 것부터 index에 할당 
        for index in range(now_position + K, now_position, -1):
            if stop_arr[index]:
                charge_count += 1
                now_position = index
                break
        else:                            # for 를 다 돌았는데 충전소 없음
            charge_count = 0
            break

    print(f"#{test_case} {charge_count}")
```
